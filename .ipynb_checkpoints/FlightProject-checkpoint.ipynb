# Flight Analysis


```python
#Import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
# Read the Data
flightdata = pd.read_csv("data/avgaf-quarters.csv")
passengerdata = pd.read_csv("data/ps.csv")
oildata = pd.read_csv("data/oil.csv")
```

```python
passengerdata.head()
```

```python
passenger=passengerdata.rename(columns={'passengers':'Passengers','year':'Year','quarter':'Quarter'})
passenger.head()
```

```python
#Clean the passenger data
passenger.fillna(0).head()
```

```python
# Rename columns 
flight = flightdata.rename(columns={'year' : 'Year','quarter':'Quarter','airportname' : 'Airport','cityname' : 'City','statename' : 'State','avgfare' : 'AvgFare','inflationadj' : 'InflationAdj','passengersample' : 'Passenger'})
```

```python
#new dataframe
select_columns = ["Year","Quarter", "Airport", "City", "State", "AvgFare", "InflationAdj", "Passenger"]
select_columns
flight = flight[select_columns]
flight.head()
```

```python
#Clean the flight data
flight = flight.fillna(0)
flight.head()
```

```python
oildata = oildata.fillna(0)
oildata.head()
```

# Aggregation and Further Cleanup

```python
#Group flight data by year and find the means
flight_mean = flight.groupby(["Year","Quarter"]).agg({"InflationAdj":"mean"})
flight_mean
```

```python
passenger_quarterly = passenger.groupby(["Year","Quarter"]).agg({"Passengers":"sum"})
passenger_quarterly
```

```python
merged_quarterly = passenger_quarterly.merge(flight_mean,on=["Year","Quarter"])
merged_quarterly
```

```python
#Clean oildata
oildata = oildata[["Location","Quarter","Year","Value"]]
oildata = oildata.loc[oildata["Location"]=="USA"]
oildata.head()
```

```python
#Get mean of oil prices per quarter year
oilprices = oildata.groupby(["Year","Quarter"]).agg({"Value":"mean"})
oilprices
```

# Stats

```python
merged_quarterly.describe()
```

```python
#plot of passengers by quarter year. Note strong outlier in the first quarter of 2017
plt.scatter(range(len(merged_quarterly)),merged_quarterly["Passengers"])
plt.show()
plt.boxplot(merged_quarterly["Passengers"])
plt.show()
```

```python
# note p value and r value.
stats.linregress(range(len(merged_quarterly)),merged_quarterly["Passengers"])
```

```python
#Dropping outliers shown in boxplot
merged_no_outlier = merged_quarterly.reset_index(drop=False)
merged_no_outlier = merged_no_outlier[~((merged_no_outlier["Year"]==2017)&(merged_no_outlier["Quarter"]==1))]
merged_no_outlier
```

```python
#plot of passengers by quarter year Without Outliers.
plt.scatter(range(len(merged_no_outlier)),merged_no_outlier["Passengers"])
plt.show()
plt.boxplot(merged_no_outlier["Passengers"])
plt.show()
```

```python
#Some Improved Linear Regression Results without that outlier
stats.linregress(range(len(merged_no_outlier)),merged_no_outlier["Passengers"])
```

```python
#plot of inflation adjusted price by year
plt.scatter(range(len(merged_quarterly)),merged_quarterly["InflationAdj"])
```

```python
#plot of passengers by inflation adjusted fare
plt.scatter(merged_quarterly["InflationAdj"],merged_quarterly["Passengers"])
```

```python
#Plot of passengers by inflation adjusted fare without outliers
plt.scatter(merged_no_outlier["InflationAdj"],merged_no_outlier["Passengers"])
```

```python
#Linear regression on the relationship between fare and number of passengers per year. Nots ure how well this shows a trend but it may be promising.
stats.linregress(merged_no_outlier["InflationAdj"],merged_no_outlier["Passengers"])
```

```python
#plot of oil prices by year
plt.scatter(range(len(oilprices)),oilprices["Value"])
```

```python
#Since the flight data seems to lack a first quarter for 2016 we need to get rid of it in the oil data for plotting purposes.
oil_data_fixed = oilprices.reset_index(drop=False)
oil_data_fixed = oil_data_fixed[~((oil_data_fixed["Year"]==2016)&(oil_data_fixed["Quarter"]=="2016-Q1"))]
oil_data_fixed
```

```python
#plot of fare by oil prices
plt.scatter(oil_data_fixed["Value"],merged_quarterly["InflationAdj"])
```

```python
#linear regression of fare by prices
stats.linregress(oil_data_fixed["Value"],merged_quarterly["InflationAdj"])
```
